```python
# **Check the Available Models for your Google API-KEY**

# If you get a **404 Not Found error**, you need to see exactly what your API key has access to.

# Run the following code to bypasses LangChain and ask the Google system directly for your available model list.

import google.generativeai as genai
from google.colab import userdata

# Configure with your key
genai.configure(api_key=userdata.get('GOOGLE_API_KEY'))

print("Models available to your API key that support 'generateContent':")
print("-" * 60)

try:
    for m in genai.list_models():
        if 'generateContent' in m.supported_generation_methods:
            # We want the string after 'models/'
            model_id = m.name.replace('models/', '')
            print(f" {model_id}") 

except Exception as e:
    print(f" Could not list models: {e}")
```

# **Usage Summary in Google AI Studio**

* When using a Google API key for Gemini models, usage limits are primarily per project, and within that project, they are per model.

* Each model in the Gemini family has its own independent bucket of "credits" (Requests Per Minute, Tokens Per Minute, and Requests Per Day).

* If you hit the limit for gemini-1.5-pro, you can immediately switch to gemini-1.5-flash or gemini-2.0-flash because they have their own separate quotas.

* *The "Pro" models typically have much stricter limits (e.g., 5-15 RPM) compared to the "Flash" models, which are built for higher throughput.
